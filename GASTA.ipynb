 <br> <center> <font size="8" color=red> 1. The Salary For Primary Level </font> </center> <br/>



```python
def Salary_Primary():
    
    
    """
    In this step we will get salaries for Primary level
    the salary based by levels and nationality and genders and
    Year Quarter the befnifit from this step know what is minimum
    and maximum salary in the Primary level and the total salary
    also for make compare between the salaries in different levels .
    
    we will used method [loc] for access rows
    and method [agg] for Aggregate one or more 
    operations over the specified axis.
    
    """
    
    #Salary Saudi in Primry levels
    Salary_Saudi_Primary_level = df.loc[(df['Degree_Level'] == 'Primary') & (df['Nationality'] == "Saudi") & (df['Year_Quarter'])]
    SL_Saudi_Primary_Level   = Salary_Saudi_Primary_level.Salary.agg(['sum', 'min','max','mean','median'])                          
    print(f" \n \n The Count Salary For Saudi In Primary \n \n {SL_Saudi_Primary_Level}")
    
    
    #Salary Non Saudi in Primry levels
    Salary_NonSaudi_Primary_level = df.loc[(df['Degree_Level'] == 'Primary') & (df['Nationality'] == "NonSaudi")& (df['Year_Quarter'])]
    SL_NonSaudi_Primary_Level   = Salary_NonSaudi_Primary_level.Salary.agg(['sum', 'min','max','mean','median'])                          
    print(f" \n \n The Count Salary For Non Saudi In Primary \n \n {SL_NonSaudi_Primary_Level}")
    
    
    #Salary Saudi in Primry level is Male.
    Male_Saudi_Primary_Salary = df.loc[(df['Degree_Level'] == 'Primary') & (df['Nationality'] == "Saudi")&(df['Gender'] == 'Male') & (df['Year_Quarter'])]
    ML_Saudi_Primary_Level_Male   = Male_Saudi_Primary_Salary.Salary.agg(['sum', 'min','max','mean','median'])                          
    print(f" \n \n Count Salary In Primary For Saudi MAle  \n \n {ML_Saudi_Primary_Level_Male}")
    
    
    #Salary Saudi in Primry level is  Female.
    Female_Saudi_Primary_Salary = df.loc[(df['Degree_Level'] == 'Primary') & (df['Nationality'] == "Saudi")&(df['Gender'] == 'Female') & (df['Year_Quarter'])]
    Fm_Saudi_Primary_Level_Male   = Female_Saudi_Primary_Salary.Salary.agg(['sum', 'min','max','mean','median'])                          
    print(f" \n \n Count Salary In Primary For Saudi Femal  \n \n {Fm_Saudi_Primary_Level_Male}")
    
    
    #Salary Non Saudi in Primary for male.
    Male_NonSaudi_Primary_Salary = df.loc[(df['Degree_Level'] == 'Primary') & (df['Nationality'] == "NonSaudi")&(df['Gender'] == 'Male') & (df['Year_Quarter'])]
    ML_NonSaudi_Primary   = Male_NonSaudi_Primary_Salary.Salary.agg(['sum', 'min','max','mean','median'])                          
    print(f" \n \n Count Salary In Primary For Non Saudi MAle  \n \n {ML_NonSaudi_Primary}")
    
  
    #Salary  Non Saudi in Primary for female.
    Female_NonSaudi_Primary_Salary = df.loc[(df['Degree_Level'] == 'Primary') & (df['Nationality'] == "NonSaudi")&(df['Gender'] == 'Female') & (df['Year_Quarter'])]
    FM_NonSaudi_Primary   = Female_NonSaudi_Primary_Salary.Salary.agg(['sum', 'min','max','mean','median'])                          
    print(f" \n \n Count Salary In Primary For Non Saudi Female  \n \n {FM_NonSaudi_Primary}")

    
Salary_Primary()
```

```python
def Nationality_Salaries():
    
    
    """
    in this methods we will show the salry based on
    nationality , saudi and not saudi.
    
    and show the numbers all teachers based on
    nationality also.
    
    we will used method [loc] and [agg].
    and slice data.
    
    """
    
    
    # i do select the Nationality Column.
    Nationality = df.set_index("Nationality")

    # Make a list of Nationalites to subset on
    Nationality_Type = ["Saudi", "NonSaudi"]


    # Subset Nationality using .loc[]
    Nationality.loc[Nationality_Type[0]]
    Nationality.loc[Nationality_Type[1]]
    
    
    #print results
    print(f" \n \n The count Nationality Saudi in all levels is {len(Nationality.loc[Nationality_Type[0]])}")
    print(f" \n \n The count Non Nationality in all levels is {len(Nationality.loc[Nationality_Type[1]])}")
    
    
    #The num Salary for All nationality.
    print("\n")
    print(df.groupby("Nationality")[["Salary"]].agg([np.sum, np.min, np.max, np.mean, np.median]))
    
    
    
    
    
Nationality_Salaries()
```
